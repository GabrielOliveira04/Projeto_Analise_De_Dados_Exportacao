### Puxar base de dados

```python
import pandas as pd

exportacao_df = pd.read_csv("exportacoes_franca.csv")
exportacao_df = exportacao_df.loc[exportacao_df['Economic Block']=='Europe', :]
display(exportacao_df)
```

### Informações Gerais

```python
#exportacao_df.info

display(exportacao_df.describe())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
exportacao_ano = exportacao_df.groupby("Year").sum()
exportacao_ano = exportacao_ano[['US$ FOB']]
def formartar(valor):
    return f"${valor:,.2f}"

exportacao_ano["US$ FOB"] = exportacao_ano["US$ FOB"].map(formartar)
display(exportacao_ano)
```

### Quais os produtos mais exportados ao longo de todo o período?

```python
exportacao_produto = exportacao_df.groupby("SH4 Description").sum()
exportacao_produto = exportacao_produto[["US$ FOB","Net Weight"]]

exportacao_produto = exportacao_produto.sort_values("US$ FOB", ascending = False)

exportacao_produto["US$ FOB"] = exportacao_produto["US$ FOB"].map(formartar)
display(exportacao_produto)
```

### Em 2020 qual cidade mais exportou para a França?

```python
exportacao_2020 = exportacao_df.loc[exportacao_df["Year"]==2020,:]
exportacao_2020 = exportacao_2020.groupby("City").sum()
exportacao_2020 = exportacao_2020[["US$ FOB"]]
exportacao_2020 = exportacao_2020.sort_values("US$ FOB", ascending = False)
exportacao_2020["US$ FOB"] = exportacao_2020["US$ FOB"].map(formartar)

display(exportacao_2020)
```
