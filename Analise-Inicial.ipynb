### Puxar base de dados

```python
import pandas as pd

exportacao_df = pd.read_csv("exportacoes_franca.csv")
exportacao_df = exportacao_df.loc[exportacao_df['Economic Block']=='Europe', :]
display(exportacao_df)
```

### Informações Gerais

```python
#exportacao_df.info

display(exportacao_df.describe())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
exportacao_ano = exportacao_df.groupby("Year").sum()

def formartar(valor):
    return f"${valor:,.2f}"

exportacao_ano["US$ FOB"] = exportacao_ano["US$ FOB"].map(formartar)
display(exportacao_ano)
```
